### Work trough all basic plotting mechanics

```python
# let's scale it down for simplicity
ax: plt.Axes
fig, ax = plt.subplots()

data = np.arange(0, 12)
# plot data
ax.plot(data, data, label="basic data")

# always label your plot - ALWAYS!
ax.set_title("My proportional plot")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")

# render legend
ax.legend()

# configure limits of the Axis
ax.set_xlim(0, 14)
ax.set_ylim(0, 14)
```

```python
# let's add a second plot
# data = np.linspace(0, 14, 50)  # get more samples for better resolution
ax.plot(data, data*2, c="r", linestyle="--", label="data*2")  # change color and linestyle too
ax.legend()
fig
```

```python
# we can plot points without lines between them
# a plot returns the created element
ax.plot(data, data / 2, linestyle="None", marker="x", color="green", label=r"$\frac{data}{2}$")
# note the TeX notation

# more about formats can be found here under "Notes":
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot

ax.legend(loc="upper left") # refresh the legend, and move it away
fig
```

```python
# scatter allows for points to have different sizes
dot_scale = np.arange(0, data.shape[0]) * 4

# scatter is slower if all dots have the same size
ax.scatter(data, data / 4, s=dot_scale, c="orange", marker="o", label="growing dots")
ax.legend(loc="upper left")
fig
```

```python
# add certainty bars
certainty = dot_scale/60
ax.errorbar(data, data * 1.5, xerr=certainty, yerr=certainty[::-1], capsize=2, label="Weird certainty", fmt="x:c")
ax.legend()
fig
```

```python
# let's make ticks every step on y-axis
ax.set_yticks(np.arange(0, 17))  # remember we set y_lim to 14
# note: if _lim and _ticks are not set to same scale the latter one set wins!
# ax.set_ylim(0, 14) 
fig
```

```python
# make sure both axes have the same (visual) scale
ax.set_aspect("equal")
fig
```

```python
# save the plot
# make sure that we don't have unneeded empty space at borders
fig.tight_layout()
fig.savefig("well_named_plot_file.png", dpi=300)

# we could show it with fig.show() if we were in a real program and not a notebook
```

### Resolution of plotted functions
mpl doesn't assume any curvature when connecting points in a plot.  
If a curve is needed we need to have more points to connect.

```python
fig, ax = plt.subplots()
# ax.set_aspect("equal")
data = np.arange(0, np.pi*2)
ax.plot(data, np.sin(data), "x-r")  # simple formats can be entered as third arg
# the resolution doesn't represent the function...
```

```python
fig, ax = plt.subplots()
data = np.linspace(0, np.pi*2, 100)  # get a better resolution
ax.plot(data, np.sin(data), "x-g")
```

### ...and something funny

```python
import numpy as np
import matplotlib.pyplot as plt

# Activate the xkcd style
with plt.xkcd():
    # Generate some data
    x = np.linspace(0, 10, 100)
    y = np.sin(x)

    # Create the plot
    plt.plot(x, y, label='sin(x)')
    plt.title('xkcd Style Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()

    # Annotate a point with text
    plt.annotate('Maximum', xy=(np.pi / 2, 1), xytext=(3, 0.8),
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 fontsize=10,
                 )

    # Show the plot
    plt.show()
```
