```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# load the image
# read images are read-only
image = mpimg.imread("DONT_commit_me.jpeg").copy()
```

```python
# images are basically arrays of rbg values
# with the shape (y, x, 3)
# note: sometimes its (y, x, 4) if the alpha channel exists
print(type(image))
print(image.shape)
```

```python
# create a figure and axes
ax: plt.Axes
fig, ax = plt.subplots()
# turn of the axis we don't need it on a picture
ax.axis("off")  

# Display the modified image
ax.imshow(image)
```

```python
# replace the defined region with a black bar
# define the region to be replaced with a black bar
x_start = 200
x_end = 440
y_start = 400
y_end = 470

# RGB value for black is (0, 0, 0)
image[y_start:y_end, x_start:x_end, :] = [0, 0, 0]
```

```python
ax.imshow(image)

modified_image_path = "NOW_you_could_commit_me.jpg"
fig.tight_layout()
# make sure that we only save the image and no whitespace
fig.savefig(modified_image_path, bbox_inches="tight", pad_inches=0)
fig
```
